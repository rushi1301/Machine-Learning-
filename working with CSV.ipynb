If we want to convert the "Royal Challengers Bangalor" to "RCB"then 

```python
def rename(name):
    if name == "Royal Challengers Bangalor" :
        return "RCB"
    else :
        return name
```

```python
rename("Royal Challengers Bangalor")
```

now after writing the function then we have to use converter parameter 

```python
pd.read_csv('IPL Matches 2008-2020.csv' , converters={'team1':rename})
```
